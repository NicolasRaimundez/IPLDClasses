## Histogram

````python
# Histogram - plt.hist()

plt.hist(albury_weather_df["Pressure3pm"], bins=50)
plt.show()
````

````python
# Histogram - plt.hist()

plt.hist(albury_weather_df["Pressure3pm"], bins=1000)
plt.show()
````

````python
# Histogram - plt.hist()

plt.figure(figsize=(10, 5))
plt.hist(albury_weather_df["Pressure3pm"], bins=50)
plt.title("Pressure Distribution")
plt.xlabel("mbar")
plt.ylabel("Frequency")
plt.show()
````

````python
# Histogram - plt.hist()

plt.figure(figsize=(10, 5))
plt.hist(albury_weather_df["Rainfall"], bins=50)
plt.title("Rainfall Distribution")
plt.xlabel("Rainfall")
plt.ylabel("Frequency")
plt.show()
````

### Two or more histograms at the same time

````python
# WITHOUT LEGEND
plt.figure(figsize=(10, 5))
plt.hist(albury_weather_df["MaxTemp"], label="MaxTemp", bins=50, color="orange", alpha=0.5)
plt.hist(albury_weather_df["MinTemp"], label="MinTemp", bins=50, color="blue", alpha=0.5)
plt.title("Min - Max Temp Distribution")
plt.xlabel("Temperature")
plt.ylabel("Frequency")
# plt.legend()
plt.show()
````

````python
# WITH LEGEND
plt.figure(figsize=(10, 5))
plt.hist(albury_weather_df["MaxTemp"], label="MaxTemp", bins=50, color="orange", alpha=0.5)
plt.hist(albury_weather_df["MinTemp"], label="MinTemp", bins=50, color="blue", alpha=0.5)
plt.title("Min - Max Temp Distribution")
plt.xlabel("Temperature")
plt.ylabel("Frequency")
plt.legend()
plt.show()
````

````python
plt.figure(figsize=(10, 5))

plt.hist(albury_weather_df["MaxTemp"], label="MaxTemp", color="orange") #WITHOUT ALPHA AND
plt.hist(albury_weather_df["MinTemp"], label="MinTemp", color="blue")
plt.title("Min - Max Temp Distribution")
plt.xlabel("Temperature")
plt.ylabel("Frequency")
plt.legend()
plt.show()
````

````python
# Pie chart - plt.pie()

plt.figure(figsize=(10, 5))
plt.pie(albury_weather_df["WindDir3pm"].value_counts(), labels=weather_df["WindDir3pm"].value_counts().index, autopct="%1.1f%%")
plt.title("WindDir3pm")
plt.show()
````

````python
# Box plot - plt.boxplot()

plt.figure(figsize=(5, 5))
plt.boxplot(albury_weather_df["MaxTemp"].dropna())
plt.title("Max Temperature Distribution")
plt.ylabel("Max Temperature")
plt.show()
````

````python
# Second example case - Spotify data from 2023
 
# https://www.kaggle.com/datasets/nelgiriyewithana/top-spotify-songs-2023

data_path = "spotify-2023.csv"

spotify_df = pd.read_csv(data_path)
spotify_df
````

````python
# Bar chart - plt.bar()

plt.figure(figsize=(10, 5))
plt.bar(spotify_df["artist_count"].value_counts().index, spotify_df["artist_count"].value_counts().values)
plt.title("Number of Artists per Song")
plt.xlabel("Number of Artists")
plt.ylabel("Count")
plt.show()
````

````python
# Scatter plot - plt.scatter()

plt.figure(figsize=(8, 8))
plt.scatter(spotify_df["danceability_%"], spotify_df["energy_%"])
plt.title("Danceability vs Energy")
plt.xlabel("Danceability")
plt.ylabel("Energy")
plt.show()
````
